```python
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler


#https://www.geeksforgeeks.org/python-pandas-get_dummies-method/

df = pd.read_csv('D:\\testeapachebeam\\excel_files\\produtobasico.csv',sep=";", encoding='latin-1')
df.loc[:,'EAN'] = df.loc[:,'EAN'].astype(str)
df
```

```python
# Only certain columns so i don't need deal with NaN for  non
select_df = df[['SKU', 'NomeProduto', 'EstoqueAtual',
       'SaldoAtual']].dropna()
```

```python
x = pd.get_dummies(select_df.drop('EstoqueAtual', axis=1))
y = select_df.SaldoAtual

y = df.SaldoAtual

x.head()
```

```python
pipeline = make_pipeline(StandardScaler(),RandomForestRegressor())
```

```python

pipeline(steps=[('standardscaler', StandardScaler()),
('randomforestgressor', RandomForestRegressor())])
```
